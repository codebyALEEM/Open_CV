```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img,img_to_array

datagen = ImageDataGenerator(
    rotation_range =40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range =0.4,
    zoom_range=0.4,
    horizontal_flip=True,
    rescale=None,
    fill_mode='wrap' #nearest ,reflect,wrap,constant
)
```

```python
#Load the Image
img = load_img(r"C:\Users\VICTUS\Desktop\Dog_Breeds.jpg")
```

```python
#Convert to numpy array
x = img_to_array(img)
x = x.reshape((1,)+ x.shape)

#Generate augmented images and save
i=0
for batch in datagen.flow(x,batch_size=1,
                          save_to_dir=r'C:\Users\VICTUS\Desktop\open_cv\data_argument',#path location to save the images
                          save_prefix='dog',save_format='jpeg'):
    i += 1
    if i > 30:
        break #Prevent Infinite loop
```
